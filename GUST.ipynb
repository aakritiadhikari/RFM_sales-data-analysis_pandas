```python
import datetime as dt
NOW = dt.datetime(2018,11,7)
data['created_at'] = pd.to_datetime(data['created_at'])
```

```python
rfmTable = data.groupby('user_id').agg({'created_at': lambda x: (NOW - x.max()).days, 'user_id': lambda x: len(x), 'amount': lambda x: x.sum()})
rfmTable['created_at'] = rfmTable['created_at'].astype(int)
rfmTable.rename(columns={'created_at': 'recency', 
                         'user_id': 'frequency', 
                         'amount': 'monetary_value'}, inplace=True)
```

```python
rfmTable.head()
```

```python
print(rfmTable)
```

```python
rfmTable1 = rfmTable.groupby('recency').agg({'recency': lambda x: len(x), 'monetary_value': lambda x: x.sum()})
rfmTable1.rename(columns={'recency': 'frequency', 
                         'amount': 'monetary_value'}, inplace=True)
print(rfmTable1)
```

```python
rfmTable1['cum_sum_user'] = rfmTable1.frequency.cumsum()
rfmTable1['cum_sum'] = rfmTable1.monetary_value.cumsum()
rfmTable1
```

```python
rfmTable1['ARPC'] = rfmTable1['cum_sum']/rfmTable1['cum_sum_user']
rfmTable1
```

```python
prediction_df = pd.DataFrame(columns= ['ARPC'],
                             data= rfmTable1['ARPC'])
pre=prediction_df.to_csv('ARPC.csv')
```
